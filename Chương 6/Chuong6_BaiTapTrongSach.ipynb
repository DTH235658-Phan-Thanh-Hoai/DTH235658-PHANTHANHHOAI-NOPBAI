## BÀI 11

```python

import os

FILENAME = "sanpham.txt"

def nhap_thong_tin_san_pham():
    """Chức năng nhập và lưu sản phẩm vào file."""
    print("--- Thêm sản phẩm mới ---")
    try:
        # Nhập thông tin
        ma = input("Nhập mã sản phẩm: ")
        ten = input("Nhập tên sản phẩm: ")
        don_gia = float(input("Nhập đơn giá: "))
        
        # Tạo chuỗi dữ liệu theo mẫu
        # MSSP;Tên Sản phẩm;Đơn giá
        data_line = f"{ma};{ten};{don_gia}\n"
        
        # Ghi nối tiếp (append mode 'a') vào file
        with open(FILENAME, 'a', encoding='utf-8') as f:
            f.write(data_line)
            
        print(f"Đã lưu sản phẩm '{ten}' vào file {FILENAME}")

    except ValueError:
        print("Lỗi: Đơn giá phải là một con số.")
    except Exception as e:
        print(f"Có lỗi xảy ra: {e}")

def xuat_danh_sach_san_pham():
    """Chức năng a: Xuất danh sách sản phẩm từ File."""
    print("--- Danh sách sản phẩm ---")
    
    # Kiểm tra file có tồn tại và có nội dung không
    if not os.path.exists(FILENAME) or os.path.getsize(FILENAME) == 0:
        print(f"File {FILENAME} không tồn tại hoặc rỗng.")
        return

    try:
        with open(FILENAME, 'r', encoding='utf-8') as f:
            print(f"{'Mã SP':<10} | {'Tên Sản phẩm':<20} | {'Đơn giá':<10}")
            print("-" * 44)
            for line in f:
                # Bỏ qua các dòng trống (nếu có)
                if not line.strip():
                    continue
                
                # Tách chuỗi bằng dấu ';'
                parts = line.strip().split(';')
                
                if len(parts) == 3:
                    ma, ten, don_gia = parts
                    print(f"{ma:<10} | {ten:<20} | {float(don_gia):<10.1f}")
                else:
                    print(f"Dòng lỗi định dạng: {line.strip()}")
                    
    except Exception as e:
        print(f"Có lỗi xảy ra khi đọc file: {e}")

# --- Chương trình chính ---
def main_bai11():
    while True:
        print("\n--- MENU BÀI 11 ---")
        print("1. Thêm sản phẩm (Nhập thông tin)")
        print("2. Xuất danh sách sản phẩm (Chức năng a)")
        print("0. Thoát")
        
        choice = input("Nhập lựa chọn của bạn: ")
        
        if choice == '1':
            nhap_thong_tin_san_pham()
        elif choice == '2':
            xuat_danh_sach_san_pham()
        elif choice == '0':
            print("Kết thúc chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")

if __name__ == "__main__":
    main_bai11()
```

## BÀI 12

```python
FILENAME = "data_bai12.txt"

def doc_file_va_in_list():
    """
    Chức năng a: Viết hàm đọc file, mỗi dòng khởi tạo 
    thành 1 list và xuất ra màn hình.
    """
    print("--- Chức năng a: Đọc file và in List ---")
    try:
        with open(FILENAME, 'r') as f:
            for i, line in enumerate(f):
                # 1. line.strip() để xóa ký tự xuống dòng '\n'
                # 2. line.strip().split(',') để tách chuỗi tại các dấu ','
                string_list = line.strip().split(',')
                
                # 3. Chuyển đổi từng phần tử trong list string sang số (int)
                number_list = [int(item) for item in string_list]
                
                # 4. Xuất ra màn hình
                print(f"Dòng {i+1}: List = {number_list}")
                
    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy file '{FILENAME}'.")
        print("Vui lòng tạo file data_bai12.txt với nội dung như trong đề bài.")
    except ValueError:
        print("Lỗi: File chứa dữ liệu không phải là số nguyên.")
    except Exception as e:
        print(f"Có lỗi xảy ra: {e}")

def xuat_cac_so_am():
    """
    Chức năng b: Xuất các số âm trên mỗi dòng ra màn hình.
    """
    print("\n--- Chức năng b: Xuất các số âm ---")
    try:
        with open(FILENAME, 'r') as f:
            for i, line in enumerate(f):
                string_list = line.strip().split(',')
                number_list = [int(item) for item in string_list]
                
                # Lọc ra các số âm
                negative_numbers = [num for num in number_list if num < 0]
                
                if negative_numbers:
                    print(f"Dòng {i+1}: Các số âm = {negative_numbers}")
                else:
                    print(f"Dòng {i+1}: Không có số âm.")
                    
    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy file '{FILENAME}'.")
    except Exception as e:
        print(f"Có lỗi xảy ra: {e}")

# --- Chương trình chính ---
def main_bai12():
    doc_file_va_in_list()
    xuat_cac_so_am()

if __name__ == "__main__":
    main_bai12()
```

## BÀI 13

```python
import os

# --- Định nghĩa Lớp (Classes) ---
class SanPham:
    """Đại diện cho một sản phẩm."""
    def __init__(self, ma, ten, gia):
        self.ma = ma
        self.ten = ten
        self.gia = float(gia)
    
    def __str__(self):
        """Trả về chuỗi đại diện cho sản phẩm."""
        return f"\t- SP: {self.ma} - {self.ten} - (Giá: {self.gia:,.0f} đ)"

class DanhMuc:
    """Đại diện cho một danh mục, chứa nhiều sản phẩm."""
    def __init__(self, ma, ten):
        self.ma = ma
        self.ten = ten
        self.danh_sach_san_pham = [] # List các đối tượng SanPham
    
    def them_san_pham(self, san_pham):
        """Thêm một đối tượng SanPham vào danh mục."""
        self.danh_sach_san_pham.append(san_pham)
    
    def __str__(self):
        """Trả về chuỗi đại diện cho danh mục và các sản phẩm bên trong."""
        # Lấy chuỗi của từng sản phẩm và nối lại bằng '\n'
        sp_str = "\n".join(str(sp) for sp in self.danh_sach_san_pham)
        if not sp_str:
            sp_str = "\t(Chưa có sản phẩm)"
        return f"[{self.ma}] {self.ten}:\n{sp_str}"

# --- Quản lý dữ liệu toàn cục ---
# List chứa các đối tượng DanhMuc
DANH_SACH_QUAN_LY = []
FILE_DANHMUC = "danhmuc.txt"
FILE_SANPHAM = "sanpham_data.txt" # Đổi tên để không trùng bài 11

# --- Các chức năng chính ---
def them_danh_muc():
    print("--- Thêm danh mục mới ---")
    ma_dm = input("Nhập mã danh mục: ")
    ten_dm = input("Nhập tên danh mục: ")
    dm = DanhMuc(ma_dm, ten_dm)
    DANH_SACH_QUAN_LY.append(dm)
    print(f"Đã thêm danh mục '{ten_dm}'.")

def tim_danh_muc_theo_ma(ma_dm):
    """Hàm tiện ích để tìm danh mục theo mã."""
    for dm in DANH_SACH_QUAN_LY:
        if dm.ma == ma_dm:
            return dm
    return None

def them_san_pham():
    print("--- Thêm sản phẩm mới ---")
    ma_dm = input("Nhập mã danh mục chứa sản phẩm: ")
    dm = tim_danh_muc_theo_ma(ma_dm)
    
    if dm is None:
        print(f"Lỗi: Không tìm thấy danh mục có mã '{ma_dm}'.")
        return
        
    print(f"Đang thêm sản phẩm vào danh mục: {dm.ten}")
    ma_sp = input("Nhập mã sản phẩm: ")
    ten_sp = input("Nhập tên sản phẩm: ")
    gia_sp = float(input("Nhập đơn giá: "))
    
    sp = SanPham(ma_sp, ten_sp, gia_sp)
    dm.them_san_pham(sp)
    print(f"Đã thêm sản phẩm '{ten_sp}' vào danh mục '{dm.ten}'.")

def hien_thi_tat_ca():
    print("--- Toàn bộ danh mục và sản phẩm ---")
    if not DANH_SACH_QUAN_LY:
        print("Chưa có dữ liệu.")
        return
        
    for dm in DANH_SACH_QUAN_LY:
        print(dm)
        print("-" * 20)

def luu_du_lieu_ra_file():
    """
    Lưu dữ liệu vào 2 file:
    1. danhmuc.txt: Lưu các danh mục (Mã;Tên)
    2. sanpham_data.txt: Lưu các sản phẩm (Mã SP;Tên SP;Giá;Mã DM)
    """
    print("Đang lưu dữ liệu...")
    try:
        with open(FILE_DANHMUC, 'w', encoding='utf-8') as f_dm:
            with open(FILE_SANPHAM, 'w', encoding='utf-8') as f_sp:
                for dm in DANH_SACH_QUAN_LY:
                    # Ghi vào file danhmuc.txt
                    f_dm.write(f"{dm.ma};{dm.ten}\n")
                    # Ghi các sản phẩm của danh mục này vào file sanpham.txt
                    for sp in dm.danh_sach_san_pham:
                        f_sp.write(f"{sp.ma};{sp.ten};{sp.gia};{dm.ma}\n")
        print("Lưu dữ liệu thành công!")
    except Exception as e:
        print(f"Lỗi khi lưu file: {e}")

def doc_du_lieu_tu_file():
    """
    Đọc dữ liệu từ 2 file và tái tạo cấu trúc
    """
    global DANH_SACH_QUAN_LY
    DANH_SACH_QUAN_LY = []
    print("Đang tải dữ liệu...")
    
    try:
        # Bước 1: Đọc file danh mục
        if not os.path.exists(FILE_DANHMUC):
            print("Không tìm thấy file dữ liệu. (Bắt đầu với danh sách rỗng)")
            return
            
        with open(FILE_DANHMUC, 'r', encoding='utf-8') as f_dm:
            for line in f_dm:
                if not line.strip(): continue
                ma_dm, ten_dm = line.strip().split(';')
                DANH_SACH_QUAN_LY.append(DanhMuc(ma_dm, ten_dm))
                
        # Bước 2: Đọc file sản phẩm và gán vào danh mục tương ứng
        if not os.path.exists(FILE_SANPHAM):
            return # Không có file sản phẩm, chỉ cần danh mục là đủ
            
        with open(FILE_SANPHAM, 'r', encoding='utf-8') as f_sp:
            for line in f_sp:
                if not line.strip(): continue
                ma_sp, ten_sp, gia_sp, ma_dm_cua_sp = line.strip().split(';')
                
                # Tạo đối tượng sản phẩm
                sp = SanPham(ma_sp, ten_sp, float(gia_sp))
                
                # Tìm danh mục của sản phẩm này
                dm = tim_danh_muc_theo_ma(ma_dm_cua_sp)
                if dm:
                    dm.them_san_pham(sp)
                    
        print(f"Đã tải {len(DANH_SACH_QUAN_LY)} danh mục.")
    except Exception as e:
        print(f"Lỗi khi đọc file: {e}")

# --- Chương trình chính ---
def main_bai13():
    # Tải dữ liệu khi chương trình bắt đầu
    doc_du_lieu_tu_file()
    
    while True:
        print("\n--- MENU BÀI 13: QUẢN LÝ SẢN PHẨM ---")
        print("1. Thêm danh mục mới")
        print("2. Thêm sản phẩm mới vào danh mục")
        print("3. Hiển thị tất cả")
        print("4. Lưu dữ liệu (Ghi ra File)")
        print("5. Tải dữ liệu (Đọc từ File)")
        print("0. Thoát")
        
        choice = input("Nhập lựa chọn: ")
        if choice == '1':
            them_danh_muc()
        elif choice == '2':
            them_san_pham()
        elif choice == '3':
            hien_thi_tat_ca()
        elif choice == '4':
            luu_du_lieu_ra_file()
        elif choice == '5':
            doc_du_lieu_tu_file()
        elif choice == '0':
            print("Kết thúc.")
            break
        else:
            print("Lựa chọn không hợp lệ.")

if __name__ == "__main__":
    main_bai13()
```

## BÀI 14

```python
import os

# --- Định nghĩa Lớp ---
class SinhVien:
    def __init__(self, ma_sv, ho_ten, lop, que_quan):
        self.ma_sv = ma_sv
        self.ho_ten = ho_ten
        self.lop = lop
        self.que_quan = que_quan
    
    def __str__(self):
        """Dùng để in ra màn hình."""
        return f"Mã SV: {self.ma_sv}, Tên: {self.ho_ten}, Lớp: {self.lop}, Quê: {self.que_quan}"
    
    def to_file_line(self):
        """Dùng để ghi vào file."""
        return f"{self.ma_sv};{self.ho_ten};{self.lop};{self.que_quan}\n"

# --- Quản lý dữ liệu toàn cục ---
danh_sach_sv = []
FILENAME_CHUNG = "svchung.txt"

# --- Các chức năng ---
def nhap_thong_tin_sv():
    """Chức năng a: Nhập thông tin sinh viên."""
    print("--- Nhập thông tin sinh viên ---")
    ma_sv = input("Mã sinh viên: ")
    ho_ten = input("Họ tên: ")
    lop = input("Lớp: ")
    que_quan = input("Quê quán: ")
    
    sv = SinhVien(ma_sv, ho_ten, lop, que_quan)
    danh_sach_sv.append(sv)
    print("Đã thêm sinh viên.")

def luu_vao_file_chung():
    """Chức năng b: Lưu thông tin vào tệp svchung.txt."""
    print(f"Đang lưu vào {FILENAME_CHUNG}...")
    try:
        with open(FILENAME_CHUNG, 'w', encoding='utf-8') as f:
            for sv in danh_sach_sv:
                f.write(sv.to_file_line())
        print(f"Đã lưu {len(danh_sach_sv)} sinh viên vào file.")
    except Exception as e:
        print(f"Lỗi khi lưu file: {e}")

def hien_thi_theo_lop():
    """Chức năng c: Hiển thị ra màn hình danh sách sinh viên của lớp X."""
    lop_tim = input("Nhập tên lớp cần tìm (ví dụ: DHKTPM17A): ").strip()
    print(f"--- Danh sách sinh viên lớp {lop_tim} ---")
    
    count = 0
    for sv in danh_sach_sv:
        if sv.lop.lower() == lop_tim.lower():
            print(sv)
            count += 1
            
    if count == 0:
        print("Không tìm thấy sinh viên nào thuộc lớp này.")

def hien_thi_theo_ten():
    """Chức năng d: Hiển thị ra màn hình danh sách sinh viên có tên X."""
    ten_tim = input("Nhập tên sinh viên cần tìm: ").strip().lower()
    print(f"--- Kết quả tìm kiếm cho tên '{ten_tim}' ---")
    
    count = 0
    for sv in danh_sach_sv:
        if ten_tim in sv.ho_ten.lower():
            print(sv)
            count += 1
    
    if count == 0:
        print("Không tìm thấy sinh viên nào có tên phù hợp.")

def doc_du_lieu_tu_file_chung():
    """Hàm tiện ích để đọc file svchung.txt vào bộ nhớ."""
    global danh_sach_sv
    danh_sach_sv = []
    
    if not os.path.exists(FILENAME_CHUNG):
        print(f"File {FILENAME_CHUNG} không tồn tại. Bỏ qua.")
        return

    print(f"Đang đọc dữ liệu từ {FILENAME_CHUNG}...")
    try:
        with open(FILENAME_CHUNG, 'r', encoding='utf-8') as f:
            for line in f:
                if not line.strip(): continue
                ma_sv, ho_ten, lop, que_quan = line.strip().split(';')
                sv = SinhVien(ma_sv, ho_ten, lop, que_quan)
                danh_sach_sv.append(sv)
        print(f"Đã tải {len(danh_sach_sv)} sinh viên.")
    except Exception as e:
        print(f"Lỗi khi đọc file: {e}")

def tach_file_theo_lop():
    """
    Chức năng e: Đọc từ svchung.txt và ghi ra các file riêng theo tên lớp.
    """
    # Đảm bảo danh_sach_sv được cập nhật từ file
    doc_du_lieu_tu_file_chung()
    
    if not danh_sach_sv:
        print("Không có dữ liệu để tách file.")
        return

    # Dùng dictionary để gom nhóm sinh viên theo lớp
    # { 'DHKTPM17A': [sv1, sv2], 'DHKTPM17B': [sv3] }
    du_lieu_theo_lop = {}
    
    for sv in danh_sach_sv:
        if sv.lop not in du_lieu_theo_lop:
            # Nếu lớp này chưa có trong dict, tạo một list rỗng cho nó
            du_lieu_theo_lop[sv.lop] = []
        # Thêm sinh viên hiện tại vào list của lớp đó
        du_lieu_theo_lop[sv.lop].append(sv)

    print("--- Bắt đầu tách file theo lớp ---")
    # Duyệt qua dict và ghi ra từng file
    for ten_lop, ds_sv_cua_lop in du_lieu_theo_lop.items():
        ten_file_lop = f"{ten_lop}.txt"
        try:
            with open(ten_file_lop, 'w', encoding='utf-8') as f:
                for sv in ds_sv_cua_lop:
                    f.write(sv.to_file_line())
            print(f"Đã ghi {len(ds_sv_cua_lop)} sinh viên vào file {ten_file_lop}")
        except Exception as e:
            print(f"Lỗi khi ghi file {ten_file_lop}: {e}")
            
    print("Tách file hoàn tất.")

# --- Chương trình chính ---
def main_bai14():
    # Tải dữ liệu cũ (nếu có) khi bắt đầu
    doc_du_lieu_tu_file_chung()
    
    while True:
        print("\n--- MENU BÀI 14: QUẢN LÝ SINH VIÊN ---")
        print("1. Nhập thông tin sinh viên (a)")
        print("2. Lưu vào file chung (b)")
        print("3. Hiển thị theo lớp (c)")
        print("4. Hiển thị theo tên (d)")
        print("5. Tách file theo lớp (e)")
        print("6. Tải lại dữ liệu từ file chung (Reload)")
        print("0. Thoát")
        
        choice = input("Nhập lựa chọn: ")
        if choice == '1':
            nhap_thong_tin_sv()
        elif choice == '2':
            luu_vao_file_chung()
        elif choice == '3':
            hien_thi_theo_lop()
        elif choice == '4':
            hien_thi_theo_ten()
        elif choice == '5':
            tach_file_theo_lop()
        elif choice == '6':
            doc_du_lieu_tu_file_chung()
        elif choice == '0':
            print("Kết thúc.")
            break
        else:
            print("Lựa chọn không hợp lệ.")

if __name__ == "__main__":
    main_bai14()
```

## BÀI 15

```python
import os

# --- Chức năng b: Viết hàm tính tiền điện ---
def tinh_tien_dien_bac_thang(so_kw):
    """
    Tính tiền điện tiêu thụ theo giá bậc thang, trả về tổng tiền (gồm VAT).
    """
    # Các mốc tiêu thụ
    MOC_1 = 100 # (0-100)
    MOC_2 = 150 # (101-150) -> 50 số
    MOC_3 = 250 # (151-250) -> 100 số
    
    # Đơn giá
    GIA_1 = 1450
    GIA_2 = 1750
    GIA_3 = 2000
    GIA_4 = 2500
    
    tien_dien = 0
    
    if so_kw <= MOC_1: # Bậc 1 (0-100)
        tien_dien = so_kw * GIA_1
    elif so_kw <= MOC_2: # Bậc 2 (101-150)
        tien_dien = (MOC_1 * GIA_1) + \
                    ((so_kw - MOC_1) * GIA_2)
    elif so_kw <= MOC_3: # Bậc 3 (151-250)
        tien_dien = (MOC_1 * GIA_1) + \
                    ((MOC_2 - MOC_1) * GIA_2) + \
                    ((so_kw - MOC_2) * GIA_3)
    else: # Bậc 4 (> 250)
        tien_dien = (MOC_1 * GIA_1) + \
                    ((MOC_2 - MOC_1) * GIA_2) + \
                    ((MOC_3 - MOC_2) * GIA_3) + \
                    ((so_kw - MOC_3) * GIA_4)
                    
    # Tính VAT 10%
    vat = tien_dien * 0.1
    
    # Tổng tiền phải trả
    tong_tien_phai_tra = tien_dien + vat
    return tong_tien_phai_tra

# --- Lớp để chứa thông tin hộ gia đình ---
class HoGiaDinh:
    def __init__(self, toa_nha, phong, chu_ho, so_kw):
        self.toa_nha = toa_nha
        self.phong = phong
        self.chu_ho = chu_ho
        self.so_kw = so_kw
        # Tự động tính tiền điện khi tạo đối tượng
        self.tien_phai_tra = tinh_tien_dien_bac_thang(self.so_kw)

# --- Quản lý dữ liệu toàn cục ---
danh_sach_cac_ho = []

# --- Chức năng a: Nhập thông tin ---
def nhap_thong_tin_ho_gia_dinh():
    """Nhập thông tin cho một hộ, bao gồm kiểm tra (validation)."""
    print("--- Nhập thông tin hộ gia đình ---")
    
    # 1. Nhập Tòa nhà (validation: không rỗng)
    while True:
        toa_nha = input("Nhập số hiệu tòa nhà (ví dụ: A1, C1): ").strip()
        if not toa_nha: # (toa_nha == "")
            print("Lỗi: Tòa nhà không được để trống. Vui lòng nhập lại.")
        else:
            break
            
    # 2. Nhập Số phòng (validation: không rỗng)
    while True:
        phong = input("Nhập số hiệu phòng (ví dụ: 305): ").strip()
        if not phong:
            print("Lỗi: Phòng không được để trống. Vui lòng nhập lại.")
        else:
            break
            
    # 3. Nhập Tên chủ hộ (validation: không rỗng)
    while True:
        chu_ho = input("Nhập tên chủ hộ: ").strip()
        if not chu_ho:
            print("Lỗi: Tên chủ hộ không được để trống. Vui lòng nhập lại.")
        else:
            break
            
    # 4. Nhập Số kW (validation: là số và không âm)
    while True:
        try:
            so_kw = float(input("Nhập số kW điện tiêu thụ trong tháng: "))
            if so_kw < 0:
                print("Lỗi: Số kW phải là số không âm. Vui lòng nhập lại.")
            else:
                break
        except ValueError:
            print("Lỗi: Đây không phải là một con số. Vui lòng nhập lại.")
            
    # Tạo đối tượng và thêm vào danh sách
    ho = HoGiaDinh(toa_nha, phong, chu_ho, so_kw)
    danh_sach_cac_ho.append(ho)
    print(f"Đã thêm hộ gia đình ông/bà {chu_ho}, phòng {phong}, tòa {toa_nha}.")

# --- Chức năng c: Hiển thị tất cả ---
def hien_thi_danh_sach():
    """Hiển thị tất cả các hộ gia đình theo định dạng bảng."""
    print("\n" + "--- BẢNG KÊ TIỀN ĐIỆN CHUNG CƯ PHÚC HƯNG ---".center(70))
    print("-" * 70)
    # Định dạng tiêu đề cột: < (căn trái), số (độ rộng)
    print(f"{'Tòa nhà':<10} | {'Phòng':<10} | {'Tên chủ hộ':<20} | {'Số điện tiêu thụ':<15} | {'Tiền điện phải trả':<15}")
    print("=" * 70)
    
    if not danh_sach_cac_ho:
        print("Chưa có dữ liệu hộ gia đình nào.")
        return
        
    for ho in danh_sach_cac_ho:
        # :<10    (căn trái, 10 ký tự)
        # :<15.1f (căn trái, 15 ký tự, 1 chữ số thập phân, kiểu float)
        # :>15,.0f (căn phải, 15 ký tự, dấu phẩy hàng ngàn, 0 chữ số thập phân)
        print(f"{ho.toa_nha:<10} | {ho.phong:<10} | {ho.chu_ho:<20} | {ho.so_kw:<15.1f} | {ho.tien_phai_tra:>15,.0f} VNĐ")
    
    print("-" * 70)

# --- Chức năng d: Ghi file ---
def ghi_file_theo_toa_nha():
    """Ghi thông tin của các hộ gia đình, gom nhóm theo file của từng tòa nhà."""
    
    if not danh_sach_cac_ho:
        print("Không có dữ liệu để ghi file.")
        return

    # Gom nhóm dữ liệu theo tòa nhà (giống bài 14)
    du_lieu_theo_toa = {}
    for ho in danh_sach_cac_ho:
        if ho.toa_nha not in du_lieu_theo_toa:
            du_lieu_theo_toa[ho.toa_nha] = []
        du_lieu_theo_toa[ho.toa_nha].append(ho)
        
    print("--- Bắt đầu ghi file theo tòa nhà ---")
    
    # Duyệt qua dict và ghi file
    for toa_nha, danh_sach_ho_cua_toa in du_lieu_theo_toa.items():
        filename = f"{toa_nha}.txt"
        try:
            with open(filename, 'w', encoding='utf-8') as f:
                f.write(f"--- DANH SÁCH THU TIỀN ĐIỆN TÒA {toa_nha} ---\n")
                f.write("=" * 60 + "\n")
                f.write(f"{'Phòng':<10} | {'Chủ hộ':<25} | {'Số kW':<10} | {'Tiền trả (VNĐ)':<15}\n")
                f.write("-" * 60 + "\n")
                
                tong_tien_toa = 0
                for ho in danh_sach_ho_cua_toa:
                    f.write(f"{ho.phong:<10} | {ho.chu_ho:<25} | {ho.so_kw:<10.1f} | {ho.tien_phai_tra:>15,.0f}\n")
                    tong_tien_toa += ho.tien_phai_tra
                    
                f.write("=" * 60 + "\n")
                f.write(f"Tổng cộng tòa {toa_nha}: {tong_tien_toa:,.0f} VNĐ\n")
            
            print(f"Đã ghi dữ liệu của {len(danh_sach_ho_cua_toa)} hộ vào file {filename}")
            
        except Exception as e:
            print(f"Lỗi khi ghi file {filename}: {e}")
            
    print("Ghi file hoàn tất.")

# --- Chương trình chính ---
def main_bai15():
    while True:
        print("\n--- MENU BÀI 15: QUẢN LÝ TIỀN ĐIỆN ---")
        print("1. Nhập thông tin hộ gia đình (a)")
        print("2. Hiển thị tất cả hộ gia đình (c)")
        print("3. Ghi dữ liệu ra file theo tòa nhà (d)")
        print("4. [Kiểm tra] Thử hàm tính tiền điện (b)")
        print("0. Thoát")
        
        choice = input("Nhập lựa chọn: ")
        
        if choice == '1':
            nhap_thong_tin_ho_gia_dinh()
        elif choice == '2':
            hien_thi_danh_sach()
        elif choice == '3':
            ghi_file_theo_toa_nha()
        elif choice == '4':
            # Chức năng thêm để test hàm (b)
            try:
                test_kw = float(input("Nhập số kW để kiểm tra: "))
                tien = tinh_tien_dien_bac_thang(test_kw)
                print(f"Với {test_kw} kW, số tiền phải trả (gồm VAT) là: {tien:,.0f} VNĐ")
            except ValueError:
                print("Vui lòng nhập số.")
        elif choice == '0':
            print("Kết thúc.")
            break
        else:
            print("Lựa chọn không hợp lệ.")

if __name__ == "__main__":
    # Bạn có thể đổi tên hàm main_baiXX() ở đây để chạy bài tập tương ứng
    # main_bai11()
    # main_bai12()
    # main_bai13()
    # main_bai14()
    main_bai15()
```
