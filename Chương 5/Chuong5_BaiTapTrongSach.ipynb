## BÀI 1

```python
# Bài 1: Lớp TamGiac
import math

class TamGiac:
    def __init__(self, a, b, c, mau_sac="Không rõ"):
        self.a = a
        self.b = b
        self.c = c
        self.mau_sac = mau_sac

    def chu_vi(self):
        return self.a + self.b + self.c

    def dien_tich(self):
        p = self.chu_vi() / 2
        return math.sqrt(p*(p-self.a)*(p-self.b)*(p-self.c))

    def loai_tam_giac(self):
        if self.a == self.b == self.c:
            return "Tam giác đều"
        elif self.a == self.b or self.b == self.c or self.a == self.c:
            return "Tam giác cân"
        elif self.a**2 + self.b**2 == self.c**2 or self.b**2 + self.c**2 == self.a**2 or self.a**2 + self.c**2 == self.b**2:
            return "Tam giác vuông"
        else:
            return "Tam giác thường"

    def hien_thi(self):
        print(f"Tam giác màu {self.mau_sac}, loại: {self.loai_tam_giac()}, chu vi: {self.chu_vi()}, diện tích: {self.dien_tich():.2f}")

# Ví dụ:
tg = TamGiac(3, 4, 5, "Đỏ")
tg.hien_thi()
```

## BÀI 2

```python
# Bài 2: Lớp HinhChuNhat
class HinhChuNhat:
    def __init__(self, dai, rong, mau_sac="Không rõ"):
        self.dai = dai
        self.rong = rong
        self.mau_sac = mau_sac

    def chu_vi(self):
        return 2 * (self.dai + self.rong)

    def dien_tich(self):
        return self.dai * self.rong

    def hien_thi(self):
        print(f"Hình chữ nhật màu {self.mau_sac}, chu vi: {self.chu_vi()}, diện tích: {self.dien_tich()}")

hcn = HinhChuNhat(5, 3, "Xanh")
hcn.hien_thi()
```

## BÀI 3

```python
# Bài 3: Lớp HinhTron
class HinhTron:
    def __init__(self, r, mau_sac="Không rõ"):
        self.r = r
        self.mau_sac = mau_sac

    def chu_vi(self):
        return 2 * math.pi * self.r

    def dien_tich(self):
        return math.pi * self.r ** 2

    def hien_thi(self):
        print(f"Hình tròn màu {self.mau_sac}, chu vi: {self.chu_vi():.2f}, diện tích: {self.dien_tich():.2f}")

ht = HinhTron(4, "Vàng")
ht.hien_thi()
```

## BÀI 4

```python
# Bài 4: Lớp DaGiac, HinhBinhHanh, HinhVuong
class DaGiac:
    def __init__(self, so_canh, do_dai):
        self.so_canh = so_canh
        self.do_dai = do_dai

class HinhBinhHanh(DaGiac):
    def __init__(self, dai, rong):
        super().__init__(4, [dai, rong, dai, rong])

    def chu_vi(self):
        return 2 * (self.do_dai[0] + self.do_dai[1])

    def dien_tich(self, chieu_cao):
        return self.do_dai[0] * chieu_cao

class HinhVuong(HinhBinhHanh):
    def __init__(self, canh):
        super().__init__(canh, canh)

hv = HinhVuong(5)
print("Chu vi hình vuông:", hv.chu_vi())
print("Diện tích hình vuông:", hv.dien_tich(5))
```

## BÀI 5

```python
# Bài 5: Lớp Diem
import math

class Diem:
    def __init__(self, x=0, y=0, mau_sac="đen"):
        self.x = x
        self.y = y
        self.mau_sac = mau_sac

    def hien_thi(self):
        print(f"Điểm ({self.x}, {self.y}), màu {self.mau_sac}")

    def tinh_tien(self, dx, dy):
        self.x += dx
        self.y += dy

    def khoang_cach(self):
        return math.sqrt(self.x**2 + self.y**2)

d = Diem(3, 4, "xanh")
d.hien_thi()
print("Khoảng cách tới gốc tọa độ:", d.khoang_cach())
```

## BÀI 6

```python
# Bài 6: Lớp ToanHoc
class ToanHoc:
    def tinhTong(self, *nso):
        return sum(nso)

    def tinhTrungBinh(self, *nso):
        return sum(nso)/len(nso)

    def timMax(self, *nso):
        return max(nso)

    def timMin(self, *nso):
        return min(nso)

    def hienThi(self, *nso):
        print("Các số:", nso)

th = ToanHoc()
th.hienThi(3, 5, 9)
print("Tổng:", th.tinhTong(3, 5, 9))
```

## BÀI 7

```python
# Bài 7: Lớp SoHoc
class SoHoc:
    def __init__(self, number1=0, number2=0):
        self.number1 = number1
        self.number2 = number2

    def inputInfo(self):
        self.number1 = float(input("Nhập số thứ nhất: "))
        self.number2 = float(input("Nhập số thứ hai: "))

    def printInfo(self):
        print(f"number1 = {self.number1}, number2 = {self.number2}")

    def addition(self): return self.number1 + self.number2
    def subtract(self): return self.number1 - self.number2
    def multi(self): return self.number1 * self.number2
    def division(self): return self.number1 / self.number2 if self.number2 != 0 else None

sh = SoHoc(8, 4)
print("Tổng:", sh.addition())
```

## BÀI 8

```python
# Bài 8: Lớp NhanVien
class NhanVien:
    def __init__(self, ten, tuoi, dia_chi, tien_luong, tong_gio):
        self.ten = ten
        self.tuoi = tuoi
        self.dia_chi = dia_chi
        self.tien_luong = tien_luong
        self.tong_gio = tong_gio

    def tinhThuong(self):
        if self.tong_gio >= 200:
            return self.tien_luong * 0.2
        elif self.tong_gio >= 100:
            return self.tien_luong * 0.1
        else:
            return 0

    def printInfo(self):
        print(f"Tên: {self.ten}, Tuổi: {self.tuoi}, Lương: {self.tien_luong}, Thưởng: {self.tinhThuong()}")

nv = NhanVien("An", 25, "AG", 10000000, 210)
nv.printInfo()
```

## BÀI 9

```python
# Bài 9: Lớp Student
class Student:
    def __init__(self, ma_sv, diem_tb, tuoi):
        self.ma_sv = ma_sv
        self.diem_tb = diem_tb
        self.tuoi = tuoi

    def showInfo(self):
        print(f"MSSV: {self.ma_sv}, Điểm TB: {self.diem_tb}, Tuổi: {self.tuoi}")

    def hoc_bong(self):
        return self.diem_tb > 8.0

sv = Student("SV001", 8.5, 20)
sv.showInfo()
print("Có học bổng:", sv.hoc_bong())
```

## BÀI 10

```python
# Bài 10: Lớp MonHoc
class MonHoc:
    def __init__(self, ma, ten, so_tin):
        self.ma = ma
        self.ten = ten
        self.so_tin = so_tin

class QuanLyMonHoc:
    def __init__(self):
        self.ds = []

    def them(self, mh):
        self.ds.append(mh)

    def xoa(self, ma):
        self.ds = [m for m in self.ds if m.ma != ma]

    def hien_thi(self):
        for m in self.ds:
            print(m.ma, m.ten, m.so_tin)

ql = QuanLyMonHoc()
ql.them(MonHoc("MH01", "Python", 3))
ql.hien_thi()
```
